# **Combine Intersections and Street Segments into a Single Feature**

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

# Assuming df is your DataFrame and it's already been loaded
# Initialize the MinMaxScaler
scaler = MinMaxScaler()

# Normalizing the three variables: 'Number of Intersections', 'Number of Street Segments', and 'Average Street Segment Length'
df_normalized_rng = scaler.fit_transform(df[['Number of Intersections', 'Number of Street Segments']])

# Convert the normalized data back to a DataFrame
df_normalized_rng = pd.DataFrame(df_normalized_rng, columns=['Normalized Number of Intersections', 'Normalized Number of Street Segments'])

# Assigning equal weights (1/3 each) and calculating the 'Road Network Geometry' score
equal_weight = 1/3
df_normalized_rng['Road Network Geometry'] = (equal_weight * df_normalized_rng['Normalized Number of Intersections'] +
                                              equal_weight * df_normalized_rng['Normalized Number of Street Segments'])

# Adding Facility ID for reference
df_normalized_rng['Facility ID'] = df['Facility ID']

# Selecting relevant columns for the final table
final_table = df_normalized_rng[['Facility ID', 'Road Network Geometry']]

# Displaying the first few rows of the final table
print(final_table.head())
```

```python
import pandas as pd
from scipy.stats import mannwhitneyu

# Assuming 'df' is your DataFrame and it's already loaded
# df = pd.read_csv('path_to_your_dataset.csv')  # Uncomment this line if you need to load your DataFrame

# Splitting the data into two groups based on the presence of a highway
group1 = df[df['Presence of Highway'] == 1]['Parking as percent of area']
group2 = df[df['Presence of Highway'] == 0]['Parking as percent of area']

# Performing the Mann-Whitney U test
stat, p = mannwhitneyu(group1, group2)

print('Statistics=%.3f, p=%.3f' % (stat, p))

# Interpreting the result
alpha = 0.05  # Set your significance level here
if p > alpha:
    print('Same distribution (fail to reject H0)')
else:
    print('Different distribution (reject H0)')
```

# **Perform ANOVA Analysis**

```python
import pandas as pd
from scipy.stats import f_oneway

# Load your dataset
df = pd.read_csv('/content/OSM_Streetnetwork_INDEPENDENT (1).csv')  # Make sure to replace 'your_dataset.csv' with the path to your dataset

# Select the variable you want to analyze
factor = 'Decade Built Categor7'

# Create a list comprehension to extract the Walkability Scores for each category within the factor
categories = df[factor].unique()  # Gets all unique categories for the factor
scores_by_category = [df[df[factor] == category]['Walkability Score'] for category in categories]

# Perform One-Way ANOVA
anova_result = f_oneway(*scores_by_category)

print(f"ANOVA result for {factor}: F={anova_result.statistic}, p-value={anova_result.pvalue}")

# Interpreting the result
if anova_result.pvalue < 0.05:
    print(f"There are statistically significant differences in Walkability Scores across different {factor}.")
else:
    print(f"There are no statistically significant differences in Walkability Scores across different {factor}.")
```

## **Compile all of the Averages In a Single Table**

```python
# Assuming 'df' is your main dataset
# Step 1: Filter categories with 100 or more occurrences for "Mode_Type"
filtered_categories = df['Mode_Type'].value_counts() >= 100
filtered_categories = filtered_categories[filtered_categories].index.tolist()
filtered_df = df[df['Mode_Type'].isin(filtered_categories)]

# Step 2: Calculate averages for each of the new categories, including filtering "Mode_Type"
# Assuming you're calculating averages for 'Walkability Score'
average_by_mode_and_type = filtered_df.groupby('Mode_Type')['Walkability Score'].mean().reset_index()
average_by_network_size = df.groupby('System')['Walkability Score'].mean().reset_index()
average_by_cities_served_size = df.groupby('Primary Mode')['Walkability Score'].mean().reset_index()
average_by_counties_served_size = df.groupby('Facility Type')['Walkability Score'].mean().reset_index()


# Step 3: Standardize the column names and assign categories
average_by_mode_and_type = average_by_mode_and_type.rename(columns={'Mode_Type': 'Category Name'}).assign(Category='Mode_Type')
average_by_network_size = average_by_network_size.rename(columns={'System': 'Category Name'}).assign(Category='System')
average_by_cities_served_size = average_by_cities_served_size.rename(columns={'Primary Mode': 'Category Name'}).assign(Category='Primary Mode')
average_by_counties_served_size = average_by_counties_served_size.rename(columns={'Facility Type': 'Category Name'}).assign(Category='Facility Type')

# Step 4: Concatenate all standardized DataFrames
combined_averages = pd.concat([average_by_mode_and_type, average_by_network_size, average_by_cities_served_size, average_by_counties_served_size])

# Step 5: Sort the combined DataFrame and display
combined_averages_sorted = combined_averages.sort_values(by='Walkability Score', ascending=False)

# Display the updated combined and sorted table
print("Combined Averages Table:")
print(combined_averages_sorted[['Category', 'Category Name', 'Walkability Score']].to_string())
```
