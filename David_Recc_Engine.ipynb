```python
import pandas as pd
df = pd.read_excel('Agenciesandcodes1.xlsx')

print(df)
```

```python
df1=df.fillna(0)
df1.head()
```

```python
# calculate co-occurrence matrix
x = np.array(df1)
y = np.array(df1.T)
co_matrix = np.dot(y,x)
np.fill_diagonal(co_matrix, 0)
co_matrix
```

```python
#show the co-occurance matrix as a dataframe 
df1_asint = df1.astype(int)
coocc = df1_asint.T.dot(df1_asint)
coocc
```

```python
#calculate the cosine similarity 
from sklearn.metrics.pairwise import cosine_similarity
print(cosine_similarity(coocc, coocc))
```

```python
df5=(cosine_similarity(coocc, coocc))
```

```python
df5_matrix=pd.DataFrame(df5)
```

```python
df5_matrix.to_csv("co_matrix5.csv")
```

```python
#export the cosine similarity to a .csv
df2_matrix = pd.DataFrame(coocc)
df2_matrix.to_csv("co_matrix4.csv")
```

```python
#export the co-occurance matrix to a csv
coocc.to_csv("coocc.csv")
```

```python
# Select an ALI to get recommendations for (#s correspond to columns in the co-occurence matrix )
id = 478
candidate = df2_matrix.iloc[id, :]
candidate
```
