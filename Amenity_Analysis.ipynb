## **Donut Chart of Amenities by Type**

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
file_path = '/content/OSM_amenities_COMBINED_REVISED.csv'  # Replace 'path_to_your_file.csv' with the actual path to your CSV file
data = pd.read_csv(file_path)

# Prepare data for the donut chart, including "amenity=ice_cream"
existing_amenity_columns = [
    'amenity=community_centre', 'amenity=library', 'tourism=museum',
    'amenity=theatre', 'amenity=cinema', 'amenity=nightclub',
    'amenity=pharmacy', 'amenity=hospital', 'amenity=clinic',
    'amenity=doctors', 'amenity=dentist', 'shop=supermarket',
    'shop=clothes', 'amenity=place_of_worship', 'Amenity=kindergarten',
    'amenity=ice_cream'  # Ensure to include "amenity=ice_cream"
]
amenities_data = data[existing_amenity_columns].sum()

# Assign different colors for each amenity
colors = plt.cm.tab20(range(len(amenities_data)))

# Construct the donut chart with distinct colors
plt.figure(figsize=(8, 8))
plt.pie(amenities_data, labels=amenities_data.index, autopct='%1.1f%%', startangle=140, colors=colors)
plt.gca().add_artist(plt.Circle((0,0),0.70,fc='white'))  # This creates the donut shape
plt.title('Share of Amenity Categories Including Ice Cream')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load and prepare data (this part would be similar to the previous steps)
# ...

# Sorting the amenities data for consistent plotting
sorted_amenities_data = amenities_data.sort_values(ascending=False)

# Calculate percentages for labels
total_amenities = sorted_amenities_data.sum()
percentages = (sorted_amenities_data / total_amenities) * 100

# Define colors for the bars
colors = ['blue' if amenity != 'amenity=kindergarten' else 'red' for amenity in sorted_amenities_data.index]

plt.figure(figsize=(14, 8))
bars = plt.bar(sorted_amenities_data.index, sorted_amenities_data, color=colors)

# Remove grid lines
plt.grid(False)

# Adding percentage labels on top of each bar
for bar, percentage in zip(bars, percentages):
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 0.5, f'{percentage:.2f}%', ha='center', va='bottom')

plt.title('Total Amenities by Category with Percentages')
plt.xlabel('Amenity Category')
plt.ylabel('Total Count')
plt.xticks(rotation=45, ha='right')
plt.show()
```

## **Comparing Amenities with Walkability Score**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
import numpy as np

# Load the data
file_path = '/content/indepdenent_variables_amenities.csv'  # Replace with the actual path to your CSV file
data = pd.read_csv(file_path)

# Extracting the relevant data
x = data['Walk Score'].values.reshape(-1, 1)
y = data['Total Amenities'].values

# Fitting the linear regression model
model = LinearRegression().fit(x, y)

# Predictions for the regression line
y_pred = model.predict(x)

# Calculating R^2 score
r2 = r2_score(y, y_pred)

# Plotting
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='Walk Score', y='Total Amenities', alpha=0.6)
plt.plot(x, y_pred, color='red', label=f'Regression line (RÂ² = {r2:.2f})')  # Regression line
plt.title('Total Walkability Score vs Total Amenities')
plt.xlabel('Total Walkability Score')
plt.ylabel('Total Amenities')
plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

# Load the data
file_path = '/content/indepdenent_variables_amenities.csv'  # Replace with the actual path to your CSV file
data = pd.read_csv(file_path)

# Filtering out rows where "% Area as Buildings" is greater than 1
data_filtered = data[data['% Building'] <= 1]

# Extracting the relevant data after filtering
x_area_filtered = data_filtered['% Building'].values.reshape(-1, 1)
y_amenities_filtered = data_filtered['Total Amenities'].values

# Fitting the linear regression model for the filtered variables
model_area_amenities_filtered = LinearRegression().fit(x_area_filtered, y_amenities_filtered)

# Predictions for the regression line after filtering
y_pred_area_amenities_filtered = model_area_amenities_filtered.predict(x_area_filtered)

# Calculating R^2 score for the filtered variables
r2_area_amenities_filtered = r2_score(y_amenities_filtered, y_pred_area_amenities_filtered)

# Plotting the new scatter plot after filtering
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data_filtered, x='% Building', y='Total Amenities', alpha=0.6)
plt.plot(x_area_filtered.flatten(), y_pred_area_amenities_filtered, color='red', label=f'Regression line (RÂ² = {r2_area_amenities_filtered:.2f})')  # Regression line
plt.title('% Area as Buildings (â‰¤ 1) vs Total Amenities')
plt.xlabel('% Area as Buildings')
plt.ylabel('Total Amenities')
plt.legend()
plt.show()
```

# **Comparing Total Amenities with Walkability Score**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
import numpy as np

# Load the data
file_path = '/content/Independent variables and all amenities.csv'  # Replace with the actual path to your CSV file
data = pd.read_csv(file_path)

# Extracting the relevant data
x = data['Total Walkability Score'].values.reshape(-1, 1)
y = data['Total Amenities'].values

# Fitting the linear regression model
model = LinearRegression().fit(x, y)

# Predictions for the regression line
y_pred = model.predict(x)

# Calculating R^2 score
r2 = r2_score(y, y_pred)

# Plotting
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='Total Walkability Score', y='Total Amenities', alpha=0.6)
plt.plot(x, y_pred, color='red', label=f'Regression line (RÂ² = {r2:.2f})')  # Regression line
plt.title('Total Walkability Score vs Total Amenities')
plt.xlabel('Total Walkability Score')
plt.ylabel('Total Amenities')
plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.preprocessing import MinMaxScaler

# Load the data
file_path = '/content/Independent variables and all amenities.csv'  # Replace with the actual path to your CSV file
data = pd.read_csv(file_path)

# Normalizing the features
scaler = MinMaxScaler()
normalized_features = scaler.fit_transform(data[['Total Walkability Score', 'Total Amenities']])
normalized_df = pd.DataFrame(normalized_features, columns=['Normalized Walkability Score', 'Normalized Total Amenities'])

# Fitting the linear regression model on the normalized data
x_norm = normalized_df['Normalized Walkability Score'].values.reshape(-1, 1)
y_norm = normalized_df['Normalized Total Amenities'].values
model_norm = LinearRegression().fit(x_norm, y_norm)

# Predictions for the regression line on normalized data
y_pred_norm = model_norm.predict(x_norm)

# Calculating R^2 score for normalized data
r2_norm = r2_score(y_norm, y_pred_norm)

# Plotting the normalized data
plt.figure(figsize=(10, 6))
sns.scatterplot(data=normalized_df, x='Normalized Walkability Score', y='Normalized Total Amenities', alpha=0.6)
plt.plot(x_norm, y_pred_norm, color='red', label=f'Regression line (RÂ² = {r2_norm:.2f})')  # Regression line for normalized data
plt.title('Normalized Total Walkability Score vs Normalized Total Amenities')
plt.xlabel('Normalized Total Walkability Score')
plt.ylabel('Normalized Total Amenities')
plt.legend()
plt.show()
```

## **Comparing Building Density with Total Amenities**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

# Load the data
file_path = '/content/Independent variables and all amenities.csv'  # Replace with the actual path to your CSV file
data = pd.read_csv(file_path)

# Filtering out rows where "% Area as Buildings" is greater than 1
data_filtered = data[data['% Area as Buildings'] <= 1]

# Extracting the relevant data after filtering
x_area_filtered = data_filtered['% Area as Buildings'].values.reshape(-1, 1)
y_amenities_filtered = data_filtered['Total Amenities'].values

# Fitting the linear regression model for the filtered variables
model_area_amenities_filtered = LinearRegression().fit(x_area_filtered, y_amenities_filtered)

# Predictions for the regression line after filtering
y_pred_area_amenities_filtered = model_area_amenities_filtered.predict(x_area_filtered)

# Calculating R^2 score for the filtered variables
r2_area_amenities_filtered = r2_score(y_amenities_filtered, y_pred_area_amenities_filtered)

# Plotting the new scatter plot after filtering
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data_filtered, x='% Area as Buildings', y='Total Amenities', alpha=0.6)
plt.plot(x_area_filtered.flatten(), y_pred_area_amenities_filtered, color='red', label=f'Regression line (RÂ² = {r2_area_amenities_filtered:.2f})')  # Regression line
plt.title('% Area as Buildings (â‰¤ 1) vs Total Amenities')
plt.xlabel('% Area as Buildings')
plt.ylabel('Total Amenities')
plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.preprocessing import MinMaxScaler

# Load the data
file_path = '/content/Independent variables and all amenities.csv'  # Replace with the actual path to your CSV file
data = pd.read_csv(file_path)

# Filtering out rows where "% Area as Buildings" is greater than 1
data_filtered = data[data['% Area as Buildings'] <= 1]

# Normalizing the "Total Amenities" feature
scaler = MinMaxScaler()
data_filtered['Normalized Total Amenities'] = scaler.fit_transform(data_filtered[['Total Amenities']])

# Extracting the relevant data for plotting
x_area_unchanged = data_filtered['% Area as Buildings'].values.reshape(-1, 1)
y_amenities_normalized = data_filtered['Normalized Total Amenities'].values

# Fitting the linear regression model for the unchanged area and normalized amenities
model_area_amenities_norm = LinearRegression().fit(x_area_unchanged, y_amenities_normalized)

# Predictions for the regression line with normalized amenities
y_pred_area_amenities_norm = model_area_amenities_norm.predict(x_area_unchanged)

# Calculating R^2 score for the unchanged and normalized variables
r2_area_amenities_norm = r2_score(y_amenities_normalized, y_pred_area_amenities_norm)

# Plotting the new scatter plot with normalized amenities
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data_filtered, x='% Area as Buildings', y='Normalized Total Amenities', alpha=0.6)
plt.plot(x_area_unchanged.flatten(), y_pred_area_amenities_norm, color='red', label=f'Regression line (RÂ² = {r2_area_amenities_norm:.2f})')  # Regression line
plt.title('% Area as Buildings vs Normalized Total Amenities')
plt.xlabel('% Area as Buildings')
plt.ylabel('Normalized Total Amenities')
plt.legend()
plt.show()
```
